# Make Visuals from Batting Card

```python
batting_card.columns
```

**Top 10 Scorers**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define top scorers again
top_scorers = batting_card.groupby('fullName')['runs'].sum().sort_values(ascending=False).head(10)

# Convert to DataFrame for seaborn
top_scorers_df = top_scorers.reset_index()

# Plot with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=top_scorers_df, x='runs', y='fullName', palette='mako')

plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Total Runs')
plt


```

**Favourite Batsman Profile**

```python
player = 'Virat Kohli'  # Change as needed

player_runs = batting_card[batting_card['fullName'] == player].groupby('season')['runs'].sum()

player_runs.plot(kind='line', marker='o', figsize=(8,5), color='blue')
plt.title(f'{player} - Runs per Season')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.grid(True)
plt.show()
```

**Top 10 Striker**

```python
# Calculate strike rates
strike_df = batting_card.groupby('fullName').agg({'runs': 'sum', 'ballsFaced': 'sum'})
strike_df = strike_df[strike_df['ballsFaced'] >= 200]  # Filter for minimum balls faced
strike_df['strikeRate'] = (strike_df['runs'] / strike_df['ballsFaced']) * 100

# Get top 10 by strike rate
top_strikers = strike_df.sort_values(by='strikeRate', ascending=False).head(10)

# Reset index so seaborn can use names
top_strikers_reset = top_strikers.reset_index()

# Plot with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_strikers_reset,
    x='strikeRate',
    y='fullName',
    palette='crest'
)
plt.title('Top 10 Strike Rates (Min 200 Balls Faced)')
plt.xlabel('Strike Rate')
plt.ylabel('Batsman')
plt.grid(True)
plt.tight_layout()
plt.show()

```

**Most 4's and 6's**

```python
boundary_df = batting_card.groupby('fullName')[['fours', 'sixes']].sum().sort_values(by='fours', ascending=False).head(10)

boundary_df.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Batsmen by Fours and Sixes')
plt.xlabel('Batsman')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Boundary Type')
plt.show()
```

**Batsmen with Most Not Outs**

```python
not_out_df = batting_card[batting_card['isNotOut'] == True].groupby('fullName').size().sort_values(ascending=False).head(10)

not_out_df.plot(kind='bar', figsize=(10,6), color='purple')
plt.title('Top 10 Players with Most Not Outs')
plt.xlabel('Batsman')
plt.ylabel('Not Outs')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Visualization by Bowling Card

**Top 10 Wicket Takers**

```python
top_wicket_takers = bowling_card.groupby('fullName')['wickets'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_wicket_takers, x='wickets', y='fullName', palette='flare')
plt.title('Top 10 Wicket Takers in IPL')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
bowling_card.columns
```

**Best Economy Rates (Min 200 Balls Bowled)**

```python
eco_df = bowling_card.copy()
eco_df['overs'] = eco_df['overs'].astype(float)
eco_df = eco_df.groupby('fullName').agg({'conceded': 'sum', 'overs': 'sum'})
eco_df = eco_df[eco_df['overs'] >= 30]  # filter bowlers with decent sample size
eco_df['economy'] = eco_df['conceded'] / eco_df['overs']
best_eco = eco_df.sort_values(by='economy').head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=best_eco, x='economy', y='fullName', palette='crest')
plt.title('Top 10 Most Economical Bowlers (Min 30 Overs)')
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Most Dot Balls**

```python
top_dots = bowling_card.groupby('fullName')['dots'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_dots, x='dots', y='fullName', palette='cubehelix')
plt.title('Top 10 Bowlers with Most Dot Balls')
plt.xlabel('Dot Balls')
plt.ylabel('Bowler')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Most Maidens Bowled**

```python
top_maidens = bowling_card.groupby('fullName')['maidens'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_maidens, x='maidens', y='fullName', palette='magma')
plt.title('Top 10 Bowlers with Most Maidens')
plt.xlabel('Maidens')
plt.ylabel('Bowler')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Distribution of Match Results**

```python
result_team_heatmap = season_summary.pivot_table(
    index='winner',
    columns='result',
    values='id',
    aggfunc='count',
    fill_value=0
)

plt.figure(figsize=(12,8))
sns.heatmap(result_team_heatmap, annot=True, cmap='YlGnBu', fmt='d', linewidths=.5)
plt.title('Match Result Types by Winning Team')
plt.xlabel('Result Type')
plt.ylabel('Winning Team')
plt.tight_layout()
plt.show()


```

# Visualizations by Points Table

```python
points_table.columns
```

**Total Points per Team**

```python
points_table = points_table.sort_values(by='matchpoints', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(data=points_table, x='name', y='matchpoints', palette='viridis')
plt.title('Total Points per Team')
plt.xlabel('Team')
plt.ylabel('Points')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**Match Wins vs. Match Losses**

```python
points_table.set_index('name')[['matcheswon', 'matcheslost']].plot(kind='bar', stacked=True, figsize=(20,10), color=['green', 'red'])
plt.title('Match Wins vs Losses by Team')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.xticks(rotation=31)
plt.show()
```

**Net Run Rate (NRR) Comparison**

```python
plt.figure(figsize=(10,6))
sns.barplot(data=points_table, x='nrr', y='name', palette='coolwarm')
plt.title('Net Run Rate (NRR) of Teams')
plt.xlabel('NRR')
plt.ylabel('Team')
plt.show()
```

**Points Distribution**

```python
points_corr = points_table[['matchesplayed', 'matcheswon', 'matcheslost', 'nrr', 'matchpoints']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(points_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between IPL Stats and Match Points')
plt.show()
```

**Team Performance by Number of Matches Played**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=points_table, x='matchesplayed', y='matchpoints', hue='name', palette='tab20', s=100)
plt.title('Team Performance by Matches Played')
plt.xlabel('Matches Played')
plt.ylabel('Points')
plt.xticks(rotation=45)
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Distribution of Points, Wins, and NRR**

```python
plt.figure(figsize=(14,6))
sns.violinplot(data=points_table[['matchpoints', 'matcheswon', 'nrr']], palette='Set2')
plt.title('Distribution of Points, Wins, and NRR')
plt.ylabel('Values')
plt.xticks([0, 1, 2], ['Points', 'Matches Won', 'NRR'])
plt.show()
```
